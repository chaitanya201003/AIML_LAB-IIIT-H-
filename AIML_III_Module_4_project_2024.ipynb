## Slope & Derivatives

Creating a function for the derivative of $f(x)$ called df(x)

```python
def df(x):
    return 2 * x + 1
```

## Gradient Descent

```python
##########################
##########################

## TASK-1 : Complete the lines of code wherever marked as [REQUIRED] in this cell.

##########################
##########################

# Gradient Descent
new_x = 3
previous_x = 0
step_multiplier = 0.1
precision = 0.00001

x_list = [new_x]
slope_list = [df(new_x)]

for n in range(500):
    previous_x = new_x
    gradient = df(new_x)  # [REQUIRED] Calculate the gradient (df is the derivative function)
    new_x = new_x - step_multiplier * gradient  # [REQUIRED] Update the x value using the gradient and step_multiplier

    step_size = abs(new_x - previous_x)

    x_list.append(new_x)
    slope_list.append(df(new_x))  # [REQUIRED] Append the new slope after updating new_x

    if step_size < precision:
        print('Loop ran this many times:', n)
        break

print('Local minimum occurs at:', new_x)
print('Slope or df(x) value at this point is:', df(new_x))
print('f(x) value or cost at this point is:', f(new_x))
```

## Gradient Descent

```python
def gradient_descent(derivative_func, initial_guess, multiplier=0.02, precision=0.001,
                    max_iter=300):
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]

    for n in range(max_iter):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - multiplier * gradient

        step_size = abs(new_x - previous_x)
        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))

        if step_size < precision:
            break
    return new_x, x_list, slope_list
```

## Partial Derivatives & Symbolic Computation

## $$\frac{\partial f}{\partial x} = ## [REQUIRED]

## $$\frac{\partial f}{\partial y} = ## [REQUIRED]

To complete **TASK-9**, you need to calculate the partial derivatives of the given function \( f(x, y) \) with respect to \( x \) and \( y \).

### Step-by-Step Solution

Let's assume the function \( f(x, y) \) is the one used in the previous task:

\[ f(x, y) = \frac{1}{\exp(-x^2 - y^2) + 1} \]

### 1. **Partial Derivative with Respect to \( x \):**

To find \( \frac{\partial f}{\partial x} \), apply the chain rule:

\[
\frac{\partial f}{\partial x} = \frac{\partial}{\partial x} \left( \frac{1}{\exp(-x^2 - y^2) + 1} \right)
\]

Using the chain rule:

\[
\frac{\partial f}{\partial x} = -\frac{1}{(\exp(-x^2 - y^2) + 1)^2} \cdot \frac{\partial}{\partial x} (\exp(-x^2 - y^2))
\]

And then:

\[
\frac{\partial}{\partial x} (\exp(-x^2 - y^2)) = \exp(-x^2 - y^2) \cdot (-2x)
\]

Thus:

\[
\frac{\partial f}{\partial x} = \frac{2x \cdot \exp(-x^2 - y^2)}{(\exp(-x^2 - y^2) + 1)^2}
\]

### 2. **Partial Derivative with Respect to \( y \):**

Similarly, for \( \frac{\partial f}{\partial y} \):

\[
\frac{\partial f}{\partial y} = \frac{\partial}{\partial y} \left( \frac{1}{\exp(-x^2 - y^2) + 1} \right)
\]

Using the same method:

\[
\frac{\partial f}{\partial y} = -\frac{1}{(\exp(-x^2 - y^2) + 1)^2} \cdot \frac{\partial}{\partial y} (\exp(-x^2 - y^2))
\]

And:

\[
\frac{\partial}{\partial y} (\exp(-x^2 - y^2)) = \exp(-x^2 - y^2) \cdot (-2y)
\]

Thus:

\[
\frac{\partial f}{\partial y} = \frac{2y \cdot \exp(-x^2 - y^2)}{(\exp(-x^2 - y^2) + 1)^2}
\]

### Final Answers:

\[
\frac{\partial f}{\partial x} = \frac{2x \cdot \exp(-x^2 - y^2)}{(\exp(-x^2 - y^2) + 1)^2}
\]

\[
\frac{\partial f}{\partial y} = \frac{2y \cdot \exp(-x^2 - y^2)}{(\exp(-x^2 - y^2) + 1)^2}
\]



```python
from sympy import symbols, diff, exp

# Define the symbolic variables
a, b = symbols('x, y')

# Define the function f(x, y)
f = lambda a, b: 1 / (exp(-a**2 - b**2) + 1)

# Print the cost function
print('Our cost function f(x, y) is: ', f(a, b))

# Compute and print the partial derivative with respect to y
print('Partial derivative wrt y is: ', diff(f(a, b), b))

# Evaluate the function at x=1.8, y=1.0
print('Value of f(x, y) at x=1.8, y=1.0 is: ', f(a, b).evalf(subs={a:1.8, b:1.0}))

# Compute and print the partial derivative with respect to x at x=1.8, y=1.0
print('Value of partial derivative wrt x at x=1.8, y=1.0: ', diff(f(a, b), a).evalf(subs={a:1.8, b:1.0}))
```

## Batch Gradient Descent with SymPy

```python
# Setup
multiplier = 0.1
max_iter = 500
params = np.array([1.8, 1.0]) # initial guess

for n in range(max_iter):
    gradient_x = diff(f(a, b), a).evalf(subs={a:params[0], b:params[1]})
    gradient_y = diff(f(a, b), b).evalf(subs={a:params[0], b:params[1]})
    gradients = np.array([gradient_x, gradient_y])
    params = params - multiplier * gradients

# Results
print('Values in gradient array', gradients)
print('Minimum occurs at x value of: ', params[0])
print('Minimum occurs at y value of: ', params[1])
print('The cost is: ', f(params[0], params[1]))
```

## Partial Derivatives of MSE w.r.t. $\theta_0$ and $\theta_1$

## $$\frac{\partial MSE}{\partial \theta_0} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big)$$

## $$\frac{\partial MSE}{\partial \theta_1} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big) \big( x^{(i)} \big)$$

## MSE & Gradient Descent

```python
import numpy as np

# x values, y values, array of theta parameters (theta0 at index 0 and theta1 at index 1)
def grad(x, y, thetas):
    n = y.size

    # Calculate the gradients (slope values) for theta0 and theta1
    theta0_slope = (-2/n) * np.sum(y - (thetas[0] + thetas[1]*x))
    theta1_slope = (-2/n) * np.sum((y - (thetas[0] + thetas[1]*x)) * x)

    # Return as a numpy array
    return np.array([theta0_slope, theta1_slope])

# Example usage
x = np.array([1, 2, 3, 4])  # Example feature values
y = np.array([2, 3, 5, 7])  # Example target values
thetas = np.array([0.5, 1.0])  # Example theta values

gradients = grad(x, y, thetas)
print("Gradients:", gradients)
```
